```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

Check datatypes

```python
final_df.dtypes
```

How many records are there based on countries

```python
country_names = final_df['Country'].value_counts()
country_names
```

```python
country_names = final_df['Country'].value_counts().index.tolist()
country_names
```

```python
country_values = final_df['Country'].value_counts().values.tolist()
country_values
```

Making a pie chart

```python
plt.pie(country_values, labels=country_names)
plt.title("Country Distribution")
plt.show()
```

Top 3 countries uses Zomato

```python
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.1f%%')
plt.title('Top 3 Countries')
plt.show()
```

Observation => maximum no of records and transactions are from India followed by United States and then United Kindom

```python
# Which type of currency is used by which country?
currency_used = final_df.groupby('Country')['Currency'].unique()
currency_used
```

```python
# Which countries have online delivery options
online_delivery = final_df[final_df['Has Online delivery'] == 'Yes']['Country'].value_counts()
online_delivery
```

```python
# Group by 'Country' and 'Has Online delivery' columns and count total records
delivery_by_country = final_df.groupby(['Country', 'Has Online delivery']).size().reset_index(name='Count')

# Display the result
print(delivery_by_country)
```
